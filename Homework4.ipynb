## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
agelabels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
uniqueplayers['Age Group'] = pd.cut(uniqueplayers['Age'], bins, labels=agelabels)
uniqueplayers_group = uniqueplayers.groupby('Age Group')
print(uniqueplayers_group['Age'].count())
print('Percentages:')
print((17/576)*100)
print((22/576)*100)
print((107/576)*100)
print((258/576)*100)
print((77/576)*100)
print((52/576)*100)
print((31/576)*100)
print((12/576)*100)
agecategories = pd.DataFrame({
    'Age Group': ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+'],
    'Total Count': [17, 22, 107, 258, 77, 52, 31, 12],
    'Percentage of Players': [2.95, 3.82, 18.58, 44.79, 13.37, 9.03, 5.38, 2.08]
})
agecategories.set_index('Age Group', inplace=True)
agecategories
```
